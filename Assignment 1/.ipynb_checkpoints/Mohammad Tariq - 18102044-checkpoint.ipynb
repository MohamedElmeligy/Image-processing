# 1)BitQuantizeImage

```python
def BitQuantizeImage(img,k):
    image = np.copy(img)
    if(k < 0 or k > 8):
        print("No changes happened\nYou have to enter a number between 0 and 8")
        return image
    quantizedValue = (2**(k-8)) * img
    image = np.array(quantizedValue, dtype=np.uint8)
    
    return image
```

```python

quantizedImage = BitQuantizeImage(data.camera(),2)

fig2, (ax11,ax22) = plt.subplots(1, 2)
fig2.set_size_inches(16, 8)
ax11.imshow(img, cmap=plt.cm.gray)
ax22.imshow(quantizedImage, cmap=plt.cm.gray)
plt.show()
```

# 2)Histogram

```python
def ImageHistogram(img,k):
    frequancies = [0] * (k + 1)
    xAxis=[]

    width = (255/k) -0.5
    division = (255 // k) 

    for i in range(k + 1):
        xAxis.append(i * (255/k)) 
        
    if(k > 255 or k <= 0):
        print("You have to enter a number between 1 and 255")
        return 'inValid Histogram'
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            pixel = img[i][j]
            frequancies[pixel // division] += 1

    # plt.hist(img)
    
    return xAxis, frequancies, width

```

```python
xAxis, frequancies, width = ImageHistogram(img,255)

fig = plt.figure()
fig.set_size_inches(10, 5)
plt.bar(xAxis, frequancies, width=width, align='edge')
plt.show()
```

# 3)HistogramEqualize

```python
def HistogramEqualize(img):
    image = np.copy(img)
    x,histogram,z = ImageHistogram(img,255)
    
    maxValue = np.max(img)   
    imgSize = img.shape[0] * img.shape[1]

    newPixelsValue = [0] * len(histogram)
    
    cdf = 0
    for i in range(len(histogram)):
        cdf += histogram[i]/imgSize
        newPixelsValue[i] = round(maxValue * cdf)
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            pixelValue = img[i][j]
            image[i][j] = newPixelsValue[pixelValue]

    return image
```

```python
img = data.camera()
newImg = HistogramEqualize(img)

xAxis, frequancies, width = ImageHistogram(img,255)
xAxisEqualized, frequanciesEqualized, widthEqualized = ImageHistogram(newImg,255)

fig, (ax1,ax2) = plt.subplots(1, 2)
fig.set_size_inches(22, 5)
ax1.bar(xAxis, frequancies, width=width, align='edge')
ax2.bar(xAxisEqualized, frequanciesEqualized, width=widthEqualized, align='edge')
plt.show()

fig2, (ax11,ax22) = plt.subplots(1, 2)
fig2.set_size_inches(16, 8)
ax11.imshow(img, cmap=plt.cm.gray)
ax22.imshow(newImg, cmap=plt.cm.gray)
plt.show()
```
